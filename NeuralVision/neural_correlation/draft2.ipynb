```python
def convert_prob_list_dict(prob):
    new_prob = {}
    cnt = 0
    for char in range(4):
        for cond_char in range(4):
            ori_key = parse_key(char, cond_char)
            new_prob[ori_key] = prob[cnt]
            cnt +=1
    return new_prob

def train_reduce(X, y):
    ### not regress on diagonal elements:
    remove_id = [i*4 for i in range(4)]
    new_X = np.delete(X, remove_id, axis=0)#copy.deepcopy(X)
    new_y = np.delete(y, remove_id)#copy.deepcopy(y)
    return new_X, new_y

def gen_X_ft(raw):
    X_train = np.zeros((16,4))
    cnt = -1
    prob = copy.deepcopy(raw)
    for char in range(4):
        for cond_char in range(4):
            cnt+=1
            ori_key = parse_key(char, cond_char)
            if char == cond_char:
                #res[ori_key] = prob[ori_key]  
                continue  
            #print("we car calcuating, ", parse_key(char,cond_char))
            walk_prob = 0.0
            info_index = 0
            for walk_char in range(4):      
                if walk_char == char or walk_char == cond_char:
                    X_train[cnt, walk_char] = 0
                    continue
                walk_cond = parse_key(walk_char,cond_char)
                char_walk = parse_key(char, walk_char)
                X_train[cnt, walk_char] = prob[walk_cond] * prob[char_walk]
    return X_train

def parse_key(char, cond_char):
    return str(char) + "|"+ str(cond_char)
# def one_step(prob):
#     res = {}
#     for char in range(4):
#         for cond_char in range(4):
#             ori_key = parse_key(char, cond_char)
#             if char == cond_char:
#                 res[ori_key] = prob[ori_key]  
#                 continue  
#             #print("we car calcuating, ", parse_key(char,cond_char))
#             walk_prob = 0.0
#             for walk_char in range(4):      
#                 if walk_char == char:
#                     continue
#                 walk_cond = parse_key(walk_char,cond_char)
#                 char_walk = parse_key(char, walk_char)
#                 #print(char_walk,prob[char_walk], walk_cond,prob[walk_cond])
#                 walk_prob = walk_prob + 1.0*prob[walk_cond] * prob[char_walk]
#                 #print(walk_prob)
#             final_key = parse_key(char,cond_char)
#             res[final_key] = walk_prob 
#     return res
# def one_step_with_prob(prob,p):
#     res = {}
#     for char in range(4):
#         for cond_char in range(4):
#             ori_key = parse_key(char, cond_char)
#             if char == cond_char:
#                 res[ori_key] = prob[ori_key]  
#                 continue  
#             #print("we car calcuating, ", parse_key(char,cond_char))
#             walk_prob = 0.0
#             for walk_char in range(4):      
#                 if walk_char == char or walk_char == cond_char:
#                     continue
#                 walk_cond = parse_key(walk_char,cond_char)
#                 char_walk = parse_key(char, walk_char)
#                 #print(char_walk,prob[char_walk], walk_cond,prob[walk_cond])
#                 walk_prob = walk_prob + 1.0*prob[walk_cond] * prob[char_walk] * p[walk_char]
#                 #print(walk_prob)
#             final_key = parse_key(char,cond_char)
#             res[final_key] = walk_prob + prob[ori_key]
#     print(res)
```

```python
patient_list = [431, 435, 436, 441, 442]
for patient_id in patient_list:
    aa = np.mean(load_pickle(f"/media/yipeng/data/movie/Movie_Analysis/CNN_result/LSTM_multi_2_KLD/{patient_id}/mentest_cond_prob_clip_annotation.pkl")['patient_acc_stats'], axis=0)
    y_train = np.mean(load_pickle(f"/media/yipeng/data/movie/Movie_Analysis/CNN_result/LSTM_multi_2_KLD/{patient_id}/mentest_cond_prob_response_annotation.pkl")['patient_acc_stats'], axis=0)
    raw = convert_prob_list_dict(aa)
    X_train = gen_X_ft(raw)
    num_char = X_train.shape[1]
    X_train1, y_train1 = train_reduce(X_train, y_train)
    x = cp.Variable(num_char)
    cost = cp.sum_squares(X_train1 @ x - y_train1)
    prob = cp.Problem(cp.Minimize(cost), [0<=x, x<=1])#[np.zeros(num_char) <= x <= np.ones(num_char)])
    prob.solve()
    # Print result.
    print(f"\n Patient id {patient_id} The optimal value is", prob.value)
    print("The optimal x is")
    print(x.value)
    print("The norm of the residual is ", cp.norm(X_train1 @ x - y_train1, p=2).value)
```
