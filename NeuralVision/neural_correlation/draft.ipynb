```python
def cross_corr(X,Y):

    num_neurons = X.shape[0]
    R_xy = np.zeros((num_neurons,num_neurons))

    for i in range(num_neurons):
        for j in range(num_neurons):
            x_arr = X[i,:]
            y_arr = Y[j,:]
            x_arr_mean = np.mean(x_arr)
            y_arr_mean = np.mean(y_arr)
            x_arr_var = np.var(x_arr)
            y_arr_var = np.var(y_arr)
            R_xy[i,j] = ((np.dot(x_arr,y_arr)/(len(x_arr))) - x_arr_mean*y_arr_mean)/(np.sqrt(x_arr_var*y_arr_var))
    return R_xy
```
